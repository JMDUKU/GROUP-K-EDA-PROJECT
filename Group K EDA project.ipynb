```python
# Importing necessary Libraries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
from collections import Counter
```

```python
df = pd.read_excel(r"C:\Users\GADGET GHANA\Downloads\vgsales.xlsx")
```

```python
# Cleaning the columns of the dataset.

df['Publisher'] = df['Publisher'].str.replace('[//]',"")
df['Year'].dropna()
df=df.dropna()
df
```

```python
# correlation between Rank and Sales from the various outlets.

corr1 = df[['Rank','NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].corr()
corr1
```

```python
# Total number of Games published in each year across the period from 1980 to 2020

year_of_publishment = df.Year.value_counts().sort_values(ascending=False)
year_of_publishment
```

```python
# Years and sales generated

df.groupby(['Year'])[['NA_Sales','EU_Sales','JP_Sales','Other_Sales',
                     'Global_Sales']].sum()
```

```python
# Total sales generated by each Game across each of the sales centres(Top 10 games).

group1 = df.groupby(['Name'])[['NA_Sales','EU_Sales','JP_Sales','Other_Sales',
                     'Global_Sales']].sum()
group1.sort_values('Global_Sales',ascending=False).head(10)
```

```python
# Top ten gaming platforms in terms of games published on it.

df.Platform.value_counts().head(10)
```

```python
# Total sales generated by each gaming platform across the market centres.

platform_group = df.groupby(['Platform'])[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].sum()
platform_group.sort_values('Global_Sales',ascending=False)
```

```python
# Top ten publishers in terms of video games published

df.Publisher.value_counts().head(10)
```

```python
publisher_group = df.groupby(['Publisher'])[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].sum()
publisher_group.sort_values('Global_Sales',ascending=False)
```

```python
# Number of times a game has been published over the period from 1980 to 2020.

df.Name.value_counts().head(10)
```

```python
# Top ten games in terms of rank.

Top_ten_games = df.groupby(['Name'])['Rank'].mean().sort_values(ascending=True).head(10)
Top_ten_games
```

```python
# Number of video games published under each Genre.

df.Genre.value_counts()
```

```python
# Total sales generated by each genre over the market centres.

group = df.groupby(['Genre'])[['NA_Sales','EU_Sales','JP_Sales','Other_Sales',
                     'Global_Sales']].sum()
group
```

```python

genre_dict = dict(Counter(df['Genre']))
genre_dict
```

```python
# Number of Games published under each Genre

plt.figure(figsize=(10,5))
plt.bar(genre_dict.keys(), genre_dict.values(), color='blue')
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.title('Number of Games published under each Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
group_dict = dict(Counter(df['Year']))
group_dict
```

```python
# Total number of published in each year

plt.figure(figsize=(10,5))
plt.bar(group_dict.keys(), group_dict.values(), color='brown')
plt.xlabel('Name of Games')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
# Top ten most published game

Top_ten_games = dict(Counter(df['Name']).most_common(10))
Top_ten_games
```

```python
# Graph for the top ten most published games over the period from 1980 to 2020.

plt.figure(figsize=(10,5))
plt.bar(Top_ten_games.keys(), Top_ten_games.values(), color='brown')
plt.xlabel('Name of Games')
plt.ylabel('Frequency')
plt.title('Top 10 most published Game')
plt.xticks(rotation=90)
plt.show()
```

```python
# Graphical presentation of total sales from each sales outlets
sum_values = group.sum()

plt.figure(figsize=(9,5))
plt.bar(sum_values.index, sum_values.values, color='blue')
plt.xlabel('Sales outlets')
plt.ylabel('Frequency')
plt.title('Total Sales from each Market')
plt.xticks(rotation=45)
plt.show()
```

```python
# Graphical plotting of sales generated by each Genre
fig, ax = plt.subplots(figsize=(10,5))
group.plot(kind='bar', ax=ax,width=0.9)
ax.set_xlabel('Genre')
ax.set_ylabel('Sales from various markets')
ax.set_title('Sales Generated by each Genre')
plt.show()
```
